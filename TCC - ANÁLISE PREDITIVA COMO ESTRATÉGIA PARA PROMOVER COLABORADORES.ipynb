```python
# Renomeando as colunas para facilitar no entendimento dos dados 'treinamento'

treinamento = treinamento.rename(columns={'employee_id':'ID', 'department':'Departamento', 'region':'Região', 'education':'Educação','gender':'Gênero','recruitment_channel':'Recrutamento','no_of_trainings':'Sem treinamento','age':'Idade','previous_year_rating':'Avaliação anterior', 'length_of_service':'Tempo de serviço','awards_won?':'Prêmios ganhos?','avg_training_score':'Pontuaç. Méd. treinam.','is_promoted':'É promovido'})
treinamento.head()
```

```python
# Conta a quantidade de valores estão iguais a NULL 

treinamento.isnull().sum()
```

```python
del treinamento['ID']
treinamento
```

```python
# Importar a função SimpleImputer com o intuito de cobrir os registros 'NaN' para não termos divergencias durante a aplicação do modelo de dados. 
# Está função busca a média dos demais registros e aplica no NaN
# Foi solicitado a função SimpleImputer a imputação de dados medianos para valores float, e Imputação de dados mais frequentes para colunas strings. 

from sklearn.impute import SimpleImputer
ImputarDados_float = SimpleImputer(strategy='mean')
ImputarDados_String = SimpleImputer(strategy='most_frequent')
```

```python
# Após foi chamada a função fit_transform para preeencher os dados 'NaN' com valores medianos e frequentes. E utilizada o isnull().sum() para ter certeza que não existe divergências em nossa base. 

treinamento['Avaliação anterior'] = ImputarDados_float.fit_transform(treinamento[['Avaliação anterior']])
treinamento['Educação'] = ImputarDados_String.fit_transform(treinamento[['Educação']])

#teste['Avaliação anterior'] = ImputarDados_float.fit_transform(teste[['Avaliação anterior']])
#teste['Educação'] = ImputarDados_String.fit_transform(teste[['Educação']])

treinamento.isnull().sum()
```

```python
# Alterar os valores 0 e 1, para promovido 0 = Não e 1 = Sim. 

def define_promocao(num):
    if num == 0:
        return 'Não'
    elif num == 1:
        return 'Sim'
    
treinamento['Promovido?'] = treinamento['É promovido'].map(define_promocao)
treinamento['Promovido?']
```

```python
# Alterar as informações "Master's & above", "Bachelor's", 'Below Secondary, para Below Secondary = 0| Bachelor's = 1
# Master's & above = 2

def define_educ(educ):
    if educ == 'Below Secondary':
        return 0
    elif educ == "Bachelor's":
        return 1
    elif educ == "Master's & above":
        return 2
    
treinamento['Nível de educação'] = treinamento['Educação'].map(define_educ)
# teste['Nível de educação'] = teste['Educação'].map(define_educ)
treinamento['Nível de educação']
```

## Criação de modelo de Machine Learning

#### Regressão logistica 

```python
# Importação do modelo de regressão logistica 

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Remover a coluna 'É promovido' e aplicar na variável x 
# Acrescentar a coluna 'É promovido' e aplicar na variável y

X = treinamento.drop('É promovido', 1)
y = treinamento['É promovido']
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.3,shuffle=True, stratify=y, random_state=42)
```

```python
# Utilização da função "get_dummies" que irá considerar todos os valores categóricos em novas colunas que contém 0 e 1.

X = pd.get_dummies(X)
X
```

```python
X_teste = pd.get_dummies(X_teste)

X_teste
```

```python
# Inserção do modelo de 'Regressao logistica' para a variável.
# Realização do treinamento do modelo de regressão logistica.

regressao_logistica = LogisticRegression(solver = 'liblinear')

regressao_logistica.fit(X, y)
```

```python
Pred_RL = regressao_logistica.predict(X_teste)
Pred_RL
```

```python
# Importação das funções 'f1_score', 'recall_score', 'accuracy_score' para ter a noção da efetividade do modelo de regressão logistica.

from sklearn.metrics import f1_score, recall_score, accuracy_score, precision_score
```

```python
# Ter uma média ponderada da precisão e evocação, onde uma pontuação F1 atinge seu melhor valor em 1 e a pior pontuação em 0. 

f1_score(y,regressao_logistica.predict(X)) * 100
```

```python
# A rechamada é a proporção em que está o número de verdadeiros positivos e o número de falsos negativos. O recall é intuitivamente a capacidade do classificador de encontrar todas as amostras positivas.

recall_score(y,regressao_logistica.predict(X)) * 100
```

```python
# Utilização desta função para calcular a precisão do subconjunto: o conjunto de rótulos previsto para uma amostra deve corresponder exatamente ao conjunto correspondente de rótulos em y_true.

accuracy_score(y,regressao_logistica.predict(X)) * 100
```

```python
# Verificar a precisão do dados. 

precision_score(y,regressao_logistica.predict(X)) * 100
```

```python
# A predição realizada será aplicada na coluna 'É promovido' e informado 1 ou 0. Lembrando que 1 é Apto a receber promoção e 0 NÃO apto. 
X_teste['É promovido'] = Pred_RL
X_teste
```

### K-Nearest Neighbors (KNN)
